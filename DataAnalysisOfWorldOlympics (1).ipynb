### Importing Important Libraries

### 1. In how many cities Summer Olympics is held so far?

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
#List of only those sports where Gold medal is won
golddf = df.where(df['Medal']=='Gold').dropna()
golddf
```

```python
data1 = []

for Sport in golddf['Sport'].unique():
    data1.append([Sport , len(golddf[golddf['Sport']  == Sport])])

pd.DataFrame(data1,columns = ['Sport','Count']).sort_values(by='Count', ascending=False).head().plot(x = 'Sport', y = 'Count', xlabel ='Sport', ylabel='Number of Gold medals won', kind = 'bar', figsize = (10,5))
```

```python
pd.DataFrame(data1,columns = ['Sport','Count']).sort_values(by='Count', ascending=False).reset_index(drop=True).head()
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data6 = []
for i in df['Event'].unique():
     data6.append([i, len(df[df['Event'] == i])])
 
pd.DataFrame(data6,columns = ['Event','Total_Players']).sort_values(by = 'Total_Players', ascending = False).head().plot(x = 'Event', y = 'Total_Players', ylabel='No. of players', kind = 'bar', figsize = (10,5))
```

```python
pd.DataFrame(data6,columns = ['Event','Total_Players']).sort_values(by = 'Total_Players', ascending = False).head().reset_index(drop = True)
```
